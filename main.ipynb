```python
from nltk.corpus import sentiwordnet as swn
from nltk import pos_tag, map_tag #postag = to detect parts of speech | maptag to simplify tags
from nltk.stem import WordNetLemmatizer #ignore prefixes and postfix -> eating -> eat
from nltk.stem import PorterStemmer # same as above -> used for plural words -> shoes simply to shoe
import nltk
nltk.download('punkt') #tokenize the tweet -> this is a tweet = [this, is, a, tweet]
nltk.download('averaged_perceptron_tagger') # used to remove @hello
nltk.download('universal_tagset') #used to remove #hashtags
nltk.download('sentiwordnet') #sentimental words -> good, bad
nltk.download('wordnet') #normal english wordds
```

```python
"""**FINDING THE SENTIMENT OF EACH TWEET**"""

#inilize the wordnet and porterstemmer
lem = WordNetLemmatizer() 
pstem = PorterStemmer()
```

```python
def pos_senti(data):#takes
    li_swn=[] #sentimental words
    li_swn_pos=[] #positive senti words
    li_swn_neg=[] #negative senti words
    missing_words=[] #remaining words after checking every exception
    for i in range(len(data.index)):
        text = data.loc[i]['tweet'] #tweet at location i
        tokens = nltk.word_tokenize(text) # tokenize the words [this, is, a, tweet]
        tagged_sent = pos_tag(tokens) #identify parts of spech [verb, adjectiv]
        store_it = [(word, map_tag('en-ptb', 'universal', tag)) for word, tag in tagged_sent] #without tag assinged to word
        #print("Tagged Parts of Speech:",store_it)

        pos_total=0
        neg_total=0
        for word,tag in store_it:
            if(tag=='NOUN'):
                tag='n'
            elif(tag=='VERB'):
                tag='v'
            elif(tag=='ADJ'):
                tag='a'
            elif(tag=='ADV'):
                tag = 'r'
            else:
                tag='nothing'
            
            #run below code if not nothing
            if(tag!='nothing'):
                concat = word+'.'+tag+'.01' 
                try:
                    #NLTK -> VADER (valance aware dictionary and sentiment reasoner)
                    this_word_pos=swn.senti_synset(concat).pos_score()
                    this_word_neg=swn.senti_synset(concat).neg_score()
                    #print(word,tag,':',this_word_pos,this_word_neg)
                except Exception as e: #this is tweet if it ignore this
                    wor = lem.lemmatize(word) #eatign -> eat
                    concat = wor+'.'+tag+'.01'
                    # Checking if there's a possiblity of lemmatized word be accepted into SWN corpus
                    try:
                        this_word_pos=swn.senti_synset(concat).pos_score()
                        this_word_neg=swn.senti_synset(concat).neg_score()
                    except Exception as e:
                        wor = pstem.stem(word) #shoes -> shoe
                        concat = wor+'.'+tag+'.01'
                        # Checking if there's a possiblity of lemmatized word be accepted
                        try:
                            this_word_pos=swn.senti_synset(concat).pos_score()
                            this_word_neg=swn.senti_synset(concat).neg_score()
                        except:
                            missing_words.append(word)
                            continue
                pos_total+=this_word_pos
                neg_total+=this_word_neg
        li_swn_pos.append(pos_total)
        li_swn_neg.append(neg_total)
        
        if(pos_total!=0 or neg_total!=0):
            if(pos_total>neg_total):
                li_swn.append(1)
            else:
                li_swn.append(-1)
        else:
            li_swn.append(0)
            
    data['pos_score']=li_swn_pos
    data['neg_score']=li_swn_neg
    data['sentiment_score']=li_swn # -1, 0, 1
    return data
```

```python
"""**DOING PROCESS OF VECTORIZATION AND APPLYING THE ALGORITHM**"""

from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer

""" 
max_df is used for removing terms that appear too frequently, also known as "corpus-specific stop words". For example:

    max_df = 0.50 means "ignore terms that appear in more than 50% of the documents".
    max_df = 25 means "ignore terms that appear in more than 25 documents".
    
    min_df is used for removing terms that appear too infrequently. For example:

    min_df = 0.01 means "ignore terms that appear in less than 1% of the documents".
    min_df = 5 means "ignore terms that appear in less than 5 documents".

The default max_df is 1.0, which means "ignore terms that appear in more than 100% of the documents". Thus, the default setting does not ignore any terms.
"""

vectorizer = TfidfVectorizer (max_features=1000, min_df=7, max_df=0.8, stop_words=stopwords.words('english'))

"""
    The fit(data) method is used to compute the mean and std dev for a given feature to be used further for scaling.
    The transform(data) method is used to perform scaling using mean and std dev calculated using the .fit() method.
    The fit_transform() method does both fits and transform.
"""
new_data['tweet'] = vectorizer.fit_transform(new_data['tweet']).toarray()

new_data['tweet'].unique()

labels=new_data['sentiment_score']
print(labels)
value_to_pass=new_data.drop(columns=['sentiment_score'])
print(value_to_pass)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(value_to_pass, labels, test_size=0.2, random_state=0) # test size 80-20
```

```python
# Train the model using the training sets
model.fit(value_to_pass,labels)

predicted=list(model.predict(X_test))
```

```python
from textblob import TextBlob

def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

def getPolarity(text):
    return TextBlob(text).sentiment.polarity

#subjectivity -> check emotions|personal opionion if someone is giving info about depression that person isnt depreseed [1, -1]
#polarity -> what are the sentiments of actual tweet [positeive, negative, neutral]

# create two new columns
data['Subjectivity'] = data['tweet'].apply(getSubjectivity)
data['Polarity'] = data['tweet'].apply(getPolarity)

data
```

```python
def getAnalysis(score):
    if score < 0:
        return 'Negative'
    if score == 0:
        return 'Neutral'
    else:
        return 'Positive'

data['Analysis'] = data['Polarity'].apply(getAnalysis)

# show Data
data
```
